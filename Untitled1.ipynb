```python
x=df[["Area"]]
y=df["Price"]
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
z=model.fit(x_train,y_train)
pre=model.predict(x_test)
print(pre)
acc=r2_score(y_test,pre)
print(acc)
mse=mean_squared_error(y_test,pre)
print(mse)
```

```python
x1=df[["Area"]]
y1=df["Price"]
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
poly=PolynomialFeatures(degree=3)
x_poly_train=poly.fit_transform(x_train)
x_poly_test=poly.transform(x_test)
model2=LinearRegression()
model2.fit(x_poly_train,y_train)
pre2=model2.predict(x_poly_test)
acc2=r2_score(y_test,pre2)
print(acc2)
mse2=mean_squared_error(y_test,pre2)
print(mse2)
```

```python
import pandas as pd
data=pd.read_csv("Churn.csv")
df=pd.DataFrame(data)
df
df.isnull().sum()
df.drop_duplicates()
df["TotalCharges"].fillna(df["TotalCharges"].mean(),inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
x=df.drop("Churn",axis=1)
x=pd.get_dummies(x)
y=df["Churn"]
acc={}
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
for i in range(1,11):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    acc4=accuracy_score(y_test,pre)
    acc[i]=acc4
print(acc)
max1=max(acc,key=acc.get)
max1
 
```

```python
import pandas as pd
data=pd.read_csv("Churn.csv")
df=pd.DataFrame(data)
df=df.drop("customerID",axis=1)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
x=df.drop("Churn",axis=1)
y=df["Churn"]
x=pd.get_dummies(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=RandomForestClassifier(n_estimators=5,criterion="entropy",max_depth=4,random_state=42)
model.fit(x_train,y_train)
pre=model.predict(x_test)
acc=accuracy_score(y_test,pre)
print(acc)
```

```python
c=classification_report(y_test,pre)
print(c)
```

```python
con=confusion_matrix(y_test,pre)
print(con)
```
